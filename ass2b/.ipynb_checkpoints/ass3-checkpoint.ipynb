```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
df = pd.read_csv('Housing.csv', index_col='Unnamed: 0')
df.head()
```

```python
x = df[['lotsize','bedrooms','bathrms', 'stories', 'garagepl']].values 
y = df['price'].values
# print(x)
```

```python
def locally_weighted_regression(x0, X, Y, tau):
    xw = X.T * weighting_function(x0, X, tau)
    beta = np.linalg.pinv(xw @ X) @ xw @ Y
    return beta

def weighting_function(x0, x, tau):
    return np.exp(np.sum((x - x0) ** 2, axis=1) / (-2 * tau * tau))
```

```python
val = np.array([])
batch = 13
max_val = int(546/batch)
tau1 = 600
for i in range(0,max_val): 
#     print(locally_weighted_regression(x[batch*i+int(batch/2)],x[batch*i:batch*i+(batch-1)],y[batch*i:batch*i+(batch-1)],tau1).T.shape)
    val = np.append(val, np.dot(x[batch*i:batch*i+(batch-1)],locally_weighted_regression(x[batch*i+int(batch/2)],x[batch*i:batch*i+(batch-1)],y[batch*i:batch*i+(batch-1)],tau1).T))
error1 = np.abs(np.divide((val-y[:val.size]),y[:val.size]))
import matplotlib.pyplot as plt
plt.plot(error1)
```
