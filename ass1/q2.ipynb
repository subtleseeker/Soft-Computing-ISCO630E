```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([75, 80, 93, 65, 87, 71, 98, 68, 84, 77])
y = np.array([82, 78, 86, 72, 91, 80, 95, 72, 89, 74])
m = len(x)
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
def slope_bias(x,y):
    A = np.sum(x)
    B = np.sum(y)
    C = np.sum(np.square(x))
    D = np.sum(x*y)
    print(A,B,C,D)
    w1 = (m*D-A*B)/(m*C-A**2)
    w0 = (C*B-A*D)/(m*C-A**2)
    print(w0,w1)
    return (w0,w1)

def predictor(x,y):
    (w0,w1) = slope_bias(x,y)
    h = w0+w1*x
    return h
```

```python
h1 = predictor(x,y)
plt.plot(x,y,'rx',x,h1)
plt.show()
```

```python
h2 = predictor(y,x)
plt.plot(y,x,'rx',y,h2)
plt.show()
```

```python
(c,m) = slope_bias(x,y)
ans = m*76+c
print("If marks in SOC = 76, then expected marks in HUR = " + str(ans))
```

```python
(c,m) = slope_bias(y,x)
ans = m*95+c  
print("If marks in HUR = 95, then expected marks in SOC = " + str(ans))
```

```python
plt.plot(x,h1,'b',label="Regression of y on x")
plt.plot(h2, y,'r',label="Regression of x on y")
plt.plot(x,y,'xb',label="Coordinates (x,y)")
plt.plot(y,x,'or',label="Coordinates (y,x)")
plt.legend()
plt.show()
```
